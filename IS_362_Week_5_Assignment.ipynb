```python
import pandas as pd
```

```python
airports = pd.read_csv("https://raw.githubusercontent.com/tidyverse/nycflights13/refs/heads/main/data-raw/airports.csv")
weather = pd.read_csv("https://raw.githubusercontent.com/tidyverse/nycflights13/refs/heads/main/data-raw/weather.csv")
```

```python
airports.head()
```

```python
weather.head()
```

**Question One:**

```python
northermost = airports.sort_values(by="lat", ascending=False)
northermost.head(8)
```

The Northernmost airport in the United States is Dillant Hopkins Airport.

**QUESTION TWO:**

```python

east_coast = airports[(airports['lon'] <= -65) & (airports['lon'] >= -75)]

east_coast_sorted = east_coast.sort_values('lon', ascending=True)

east_coast_sorted.head(10)
```

The easternmost airport is WWD.

**QUESTION THREE:**

```python
weather["time_hour"] = pd.to_datetime(weather["time_hour"])

day_weather = weather[weather["time_hour"].dt.date == pd.to_datetime("2013-02-12").date()]

windiest = day_weather.groupby("origin")["wind_speed"].max().sort_values(ascending=False)
windiest
```

The airport that had the windest weather was EWR.
