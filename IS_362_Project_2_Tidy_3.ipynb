```python
import pandas as pd
```

```python
data = {
    'Team': ['Network Ops', 'App Support', 'Help Desk', 'Security'],
    'Jan_Tickets_Resolved': [320, 210, 450, 180],
    'Feb_Tickets_Resolved': [300, 230, 470, 190],
    'Mar_Tickets_Resolved': [310, 220, 460, 200]
}
```

```python
df_untidy = pd.DataFrame(data)
print("Original Data:")
print(df_untidy)
```

```python
df_tidy = pd.melt(
    df_untidy,
    id_vars=['Team'],
    value_vars=['Jan_Tickets_Resolved', 'Feb_Tickets_Resolved', 'Mar_Tickets_Resolved'],
    var_name='Month',
    value_name='Tickets_Resolved'
)
```

```python
df_tidy['Month'] = df_tidy['Month'].str.replace('_Tickets_Resolved', '')

print("\nTidy Data:")
print(df_tidy)
```

```python
summary = df_tidy.groupby('Team')['Tickets_Resolved'].describe()
print("Summary by Team:")
print(summary)
```

```python
monthly_totals = df_tidy.groupby('Month')['Tickets_Resolved'].sum()
print("\nTotal tickets resolved per month:")
print(monthly_totals)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
for team in df_tidy['Team'].unique():
    team_data = df_tidy[df_tidy['Team'] == team]
    plt.plot(team_data['Month'], team_data['Tickets_Resolved'], marker='o', label=team)

plt.title("Tickets Resolved by ITS Team per Month")
plt.xlabel("Month")
plt.ylabel("Tickets Resolved")
plt.legend()
plt.show()
```

I tidied the dataset so each row shows one teamâ€™s tickets for one month. Then, I analyzed trends and found that the Help Desk resolves the most tickets and Security the least and ticket counts were fairly stable over the three months.
